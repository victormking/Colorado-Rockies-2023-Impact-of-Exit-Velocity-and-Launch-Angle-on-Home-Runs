```python
new_COL_batting_records_2023 = new_COL_batting_records_2023.rename(columns={'game_date': 'Date'})
combined_stats_col_2023 = pd.merge(new_COL_batting_records_2023, COL_game_records_2023, on=["Date"], how="left")
combined_stats_col_2023['Games_at_Coors'] = combined_stats_col_2023['Home_Away'].map({'@' : '', 'Home' : 'Coors'})


combined_stats_col_2023.drop('Date_formatted_x', axis=1, inplace=True)
combined_stats_col_2023.drop('Date_formatted_y', axis=1, inplace=True)
combined_stats_col_2023.drop('Home_Away', axis=1, inplace=True)
combined_stats_col_2023.drop('Orig. Scheduled', axis=1, inplace=True)
combined_stats_col_2023.drop('spin_dir', axis=1, inplace=True)
combined_stats_col_2023.drop('break_angle_deprecated', axis=1, inplace=True)
combined_stats_col_2023.drop('spin_rate_deprecated', axis=1, inplace=True)
combined_stats_col_2023.drop('break_length_deprecated', axis=1, inplace=True)
combined_stats_col_2023.drop('hit_location', axis=1, inplace=True)


combined_stats_col_2023 = combined_stats_col_2023.rename(columns = {'launch_speed' : 'Exit Velocity', 'launch_angle' : 'Launch Angle', 'game_date' : 'Date'})
combined_stats_col_2023.astype({'Games_at_Coors': 'bool'})


combined_stats_col_2023.to_csv("combinedstats.csv")
```

```python
avg_mean_combined_stats_col_2023 = combined_stats_col_2023[['Games_at_Coors', 'Exit Velocity', 'Launch Angle']].groupby(['Games_at_Coors']).mean()
avg_mean_combined_stats_col_2023_visual = avg_mean_combined_stats_col_2023.copy()
avg_mean_combined_stats_col_2023_visual['Launch Angle'] = avg_mean_combined_stats_col_2023_visual['Launch Angle'].round(2)
avg_mean_combined_stats_col_2023_visual['Launch Angle'] = avg_mean_combined_stats_col_2023_visual['Launch Angle'].astype(str)
avg_mean_combined_stats_col_2023_visual['Launch Angle'] = avg_mean_combined_stats_col_2023_visual['Launch Angle'] + '°'
avg_mean_combined_stats_col_2023_visual['Exit Velocity'] = avg_mean_combined_stats_col_2023_visual['Exit Velocity'].round(2)
avg_mean_combined_stats_col_2023_visual['Exit Velocity'] = avg_mean_combined_stats_col_2023_visual['Exit Velocity'].astype(str)
avg_mean_combined_stats_col_2023_visual['Exit Velocity'] = avg_mean_combined_stats_col_2023_visual['Exit Velocity'] + ' mph'
avg_mean_combined_stats_col_2023_visual 

avg_mean_combined_stats_col_2023.to_csv("avg_combined_stats.csv")
```
